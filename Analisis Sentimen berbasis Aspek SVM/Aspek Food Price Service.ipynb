```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel("aspekfix.xlsx")
df #cetak isi dataset
```

```python
from sklearn.feature_extraction.text import TfidfTransformer, CountVectorizer 
vectorizer = CountVectorizer(decode_error="replace")
transformer = TfidfTransformer()
TFIDF = transformer.fit_transform(vectorizer.fit_transform(df['Clean Data']))
```

```python
print(TFIDF)
```

```python
labels = [] #membuat kolom labels
for data in df['Label']:#untuk data di df row kategori
    labels.append(data)
kolom = labels.pop
```

```python
ktgr = [] #membuat kolom labels
for data in df['Kategori']:#untuk data di df row kategori
    ktgr.append(data)
kolom = ktgr.pop
```

```python
#kemudian kategori diberi label
Label=[]
for i in range(len(labels)): 
    if ktgr[i] =='service' and labels[i]=='negatif':
        nilai = 4
    if ktgr[i] =='service' and labels[i]=='positif':
        nilai = 5 
    Label.append(nilai)
```

```python
from sklearn import svm #metode untuk klasifikasi aspek / kategori
from sklearn.feature_extraction.text import TfidfTransformer, CountVectorizer #untuk menghitung vsm dan tfidf #vsm =vector space model
from sklearn import metrics #metode untuk pembentukan matriks 1x1 1x2 1x3
from sklearn.metrics import accuracy_score #metod untuk pembentukan skor akurasi 
from sklearn.model_selection import KFold #metode untuk perhitungan kfold
import numpy as np 
import re #regular expression
import itertools #fungsi pengulangan tak terbatas
import matplotlib.pyplot as plt #pembentukan diagram
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC #suport vector classifier
from sklearn.metrics import classification_report
```

```python
akurasi=[]
def avg_akurasiSVM(): ##fungsi untuk rata rata akurasi one vs rest classifier
    total = 0
    for i in range(10): 
        total = total + akurasi[i]
    print("Rata-rata akurasi keseluruhan SVM Linear adalah :", total /10)
```

```python
from sklearn.svm import SVC #fungsi K-Fold Cross Validation melakukan insialisasi 10 kali iterasi
kFoldCrossValidation = KFold (n_splits = 10, random_state = 2, shuffle = True)

for latih, uji in kFoldCrossValidation.split(TFIDF, Label):

    print("-----------------------------")
    print("Jumlah data latih: ", len(latih)) #berapa jmlah data latih
    print("Jumlah data uji: ", len(uji)) #berapa jumlah data uji
    print("\nData Latih: \n", latih) #menampilkan data latih
    print("\nData Uji: \n", uji)#menampilkan data uji
    dataLatih1, dataUji1 = TFIDF[latih], TFIDF[uji]#proses inisialisasi dari masing2 data latih/uji dijadikan nilai tfidf lalu di copy ke variabel dataLatih/Uji1
    Label = np.array(Label)
    dataLatih2, dataUji2 = Label[latih], Label[uji]#

    SVM = SVC(kernel = 'linear', C = 1, degree = 3, gamma = 0.1)
    #SVM = SVC(kernel = 'linear')
    model = SVM.fit(dataLatih1, dataLatih2)
    prediksi = model.predict(dataUji1)

    print("\nHasil Prediksi SVM  : \n", prediksi)

    print("\nConfusion Matrix: \n", metrics.confusion_matrix(dataUji2, prediksi))

    akurasi.append(accuracy_score(dataUji2, prediksi))

    print("\nAkurasi SVM : ", accuracy_score(dataUji2, prediksi))
    print()\

    print(metrics.classification_report(dataUji2, prediksi,labels=[4,5])) #proses pembentukan confusin matrix
avg_akurasiSVM() #hasil rata rata multiclass
```
