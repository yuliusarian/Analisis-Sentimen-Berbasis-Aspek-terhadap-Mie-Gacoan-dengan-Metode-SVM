```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import string
import re
```

```python
tweet_sampah = pd.read_excel('sampahNew.xlsx')
print(tweet_sampah)
```

```python
tweet_sampah = pd.DataFrame(tweet_sampah[['Tweet']])
tweet_sampah
```

```python
tweet = tweet_sampah['Tweet']
```

```python
#cleaning dan case folding
def clean_text(text):
    #menghapus username twitter
    text=re.sub('@[^\s]+',' ',text)
    #menghapus https http
    text=re.sub(r"(?:\@|http?\://|https?\://|www)\S+","", text)
    #menghilangkan tanda baca
    text=text.translate(str.maketrans(' ',' ',string.punctuation))
    #mengganti karakter html menjadi tanda petik
    text=re.sub('<.*?>', ' ',text)
    #mempertimbangkan huruf dan angka
    text=re.sub('[^a-zA-Z]',' ',text)
    #mengganti line baru dengan spasi
    text=re.sub("\n"," ",text)
    #mengubah ke huruf kecil
    text= text.lower()
    #menghapus single char
    text=re.sub(r"\b[a-zA-Z]\b"," ",text)
    #memisahkan dan gabung kata
    text=' '.join(text.split())
    
    return text
```

```python
tweet_sampah['CaseFolding_Cleaning'] = [clean_text(i) for i in tweet]
tweet = tweet_sampah['CaseFolding_Cleaning']
```

```python
tweet_sampah.head(10)
```

```python
#tokenize
from nltk.tokenize import RegexpTokenizer
regexp = RegexpTokenizer(r'\w+|$[0-9]+|\S+')
tweet_sampah['Token'] = tweet_sampah['CaseFolding_Cleaning'].apply(regexp.tokenize)
```

```python
tweet_sampah.head(10)
```

```python
normalized_word = pd.read_csv("kamus_alay.csv")

normalized_word_dict={}
for index, row in normalized_word.iterrows():
    if row[0] not in normalized_word_dict:
        normalized_word_dict[row[0]] = row[1]
        
def normalized_term(document):
    return [normalized_word_dict[term] if term in normalized_word_dict else term for term in document]

tweet_sampah['Normalisasi'] = tweet_sampah['Token'].apply(normalized_term)
tweet_sampah
```
